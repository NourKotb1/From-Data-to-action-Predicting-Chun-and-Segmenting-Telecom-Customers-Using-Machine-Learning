```python
# Function to remove outliers using IQR
def remove_outliers_iqr(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df[col] = np.where(df[col] < lower_bound, lower_bound, df[col])  # Replace lower outliers
    df[col] = np.where(df[col] > upper_bound, upper_bound, df[col])  # Replace upper outliers

# Identify all numerical columns
numerical_columns = df.select_dtypes(include=['number']).columns

# Apply outlier removal for all numerical columns
for col in numerical_columns:
    remove_outliers_iqr(df, col)

print("Outliers handled for all numerical columns.")
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot size
plt.figure(figsize=(15, 8))
# Generate a boxplot for all numerical columns
sns.boxplot(data=df[numerical_columns])
# Rotate x-axis labels for better readability
plt.xticks(rotation=90)
# Set title
plt.title("Boxplot of Numerical Features After Outlier Handling")
# Show the plot
plt.show()
```

# Data Preprocessing

```python
class DataPreprocessor:
    def __init__(self, df, correlation_threshold=0.9):
        """
        Initialize the DataPreprocessor class.

        Parameters:
        - df: pandas DataFrame, the dataset to preprocess.
        - correlation_threshold: float, the threshold for dropping highly correlated features.
        """
        self.df = df.copy()
        self.correlation_threshold = correlation_threshold

    def drop_missing_values(self):
        """Remove rows with missing values."""
        self.df = self.df.dropna()
        return self.df

    def drop_irrelevant_columns(self, columns_to_drop):
        """Drop specified irrelevant columns."""
        self.df = self.df.drop(columns=columns_to_drop, errors='ignore')
        return self.df

    def drop_highly_correlated_features(self):
        """Identify and drop highly correlated numerical features."""
        num_cols = self.df.select_dtypes(include=[np.number]).columns
        corr_matrix = self.df[num_cols].corr()

        # Get the upper triangle of the correlation matrix
        upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

        # Identify features with high correlation
        to_drop = [column for column in upper.columns if any(upper[column] > self.correlation_threshold)]
        print("Columns to drop due to high correlation:", to_drop)

        self.df = self.df.drop(columns=to_drop, errors='ignore')
        return self.df

    def get_categorical_and_numerical_columns(self):
        """Retrieve categorical and numerical column names."""
        categorical = self.df.select_dtypes(include='object').columns.tolist()
        numerical = self.df.select_dtypes(include='number').columns.tolist()
        return categorical, numerical

    def check_duplicates(self):
        """Count the number of duplicate rows."""
        return self.df.duplicated().sum()

    def preprocess(self, columns_to_drop=[]):
        """Perform the complete preprocessing pipeline."""
        self.drop_missing_values()
        self.drop_irrelevant_columns(columns_to_drop)
        self.drop_highly_correlated_features()
        categorical, numerical = self.get_categorical_and_numerical_columns()
        duplicates = self.check_duplicates()

        print(f"Number of duplicate rows: {duplicates}")
        print(f"Categorical columns: {categorical}")
        print(f"Numerical columns: {numerical}")

        return self.df

# Instantiate and run the preprocessor
preprocessor = DataPreprocessor(df_renamed)
processed_df = preprocessor.preprocess(columns_to_drop=['N'])
```

# Customer Segmentation

```python
# Import Required Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.manifold import TSNE

# Features to use for clustering
clustering_features = [
    'Total number of months in service',
    'Number of active subscribers in household',
    'Current handset price',
    'Number of unique subscribers in the household',
    'Total revenue',
    'Average monthly minutes of use over the life of the customer',
    'Mean number of monthly minutes of use',
    'Average monthly revenue over the life of the customer',
    'Total number of calls over the life of the customer',
    'Median monthly revenue (charge amount)'
]

# Filter dataset to selected features
clustering_data = processed_df[clustering_features].dropna()

print(f"✅ Using {len(clustering_features)} refined features for clustering")

# Scale the features
scaler = StandardScaler()
scaled_data = scaler.fit_transform(clustering_data)

# Apply KMeans Clustering
kmeans = KMeans(n_clusters=4, random_state=42)
cluster_labels = kmeans.fit_predict(scaled_data)

# Feature Selection using Random Forest
rf = RandomForestClassifier(n_estimators=100, random_state=42)

# Train the Random Forest model on the scaled data (using cluster_labels as the target for feature importance)
rf.fit(scaled_data, cluster_labels)

# Get top features based on importance
feature_importance = sorted(zip(clustering_data.columns, rf.feature_importances_), key=lambda x: x[1], reverse=True)

# Select top 10 features
selected_features = [f[0] for f in feature_importance[:10]]

# Print selected features and top features
print(" Selected Features for Clustering:", selected_features)
print(" Top Features:", feature_importance[:10])

# Optionally, you could visualize feature importance
plt.figure(figsize=(10, 6))
plt.barh([f[0] for f in feature_importance[:10]], [f[1] for f in feature_importance[:10]], color='skyblue')
plt.xlabel('Feature Importance')
plt.title('Top 10 Features for Clustering')
plt.gca().invert_yaxis()  # To display the most important feature on top
plt.show()
```

```python
from sklearn.decomposition import PCA

# Keep 8 components
pca = PCA(n_components=8)
pca_transformed = pca.fit_transform(scaled_data)

# Check explained variance
print(" PCA Explained Variance Ratio:", pca.explained_variance_ratio_)
```

```python
# Visualizing clusters after PCA transformation
plt.figure(figsize=(8, 6))
sns.scatterplot(
    x=pca_transformed[:, 0], y=pca_transformed[:, 1], hue=cluster_labels, palette="viridis"
)
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.title("Customer Segmentation Clusters")
plt.legend(title="Cluster")
plt.show()
```

```python
# Determine Optimal k using Elbow Method
inertia = []
k_range = range(1, 11)  # Checking k values from 1 to 10

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(pca_transformed)
    inertia.append(kmeans.inertia_)

# Plot Elbow Curve
plt.figure(figsize=(8, 5))
plt.plot(k_range, inertia, marker='o', linestyle='dashed', color='blue')
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Inertia (Within-Cluster Sum of Squares)")
plt.title("Elbow Method for Optimal k")
plt.show()

print(" Check the elbow plot and select the optimal number of clusters ")
```

```python
# Optimal k determined from the elbow method
optimal_k = 2  # Adjust based on the elbow plot

# Apply K-Means Clustering
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
cluster_labels = kmeans.fit_predict(pca_transformed)

# Compute Silhouette Score
sil_score = silhouette_score(pca_transformed, cluster_labels)
print(f" Silhouette Score for {optimal_k} clusters: {sil_score:.4f}")

# Assign cluster labels to processed_df
processed_df['Cluster'] = cluster_labels

print(" Customer Segmentation Completed!")
```

```python
import numpy as np
import random
import tensorflow as tf
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense
import gym

# Define a simple Customer Retention Environment (Simulated)
class CustomerRetentionEnv(gym.Env):
    def __init__(self, df):
        self.df = df
        self.customer_index = 0
        self.state_size = df.shape[1] - 1  # Exclude 'churn' column
        self.action_size = 3  # Actions: (0 = No action, 1 = Offer discount, 2 = Improve service)
        self.current_state = self.df.iloc[self.customer_index, :-1].values  # Excluding churn label

    def reset(self):
        self.customer_index = 0
        self.current_state = self.df.iloc[self.customer_index, :-1].values
        return self.current_state

    def step(self, action):
        """Simulate Customer Response to Actions"""
        reward = 0
        churn = self.df.iloc[self.customer_index, -1]  # Get churn label

        if action == 1:  # Discount
            reward = 1 if churn == 0 else -1
        elif action == 2:  # Improve Service
            reward = 2 if churn == 0 else -2
        else:  # No Action
            reward = -1 if churn == 1 else 0

        self.customer_index += 1
        done = self.customer_index >= len(self.df) - 1
        if done:
            return self.current_state, reward, done, {}

        self.current_state = self.df.iloc[self.customer_index, :-1].values
        return self.current_state, reward, done, {}

```

```python
from collections import deque
import random

class DQN_Agent:
    def __init__(self, state_size, action_size):
        self.state_size = state_size
        self.action_size = action_size
        self.memory = deque(maxlen=2000)
        self.gamma = 0.95  # Discount rate
        self.epsilon = 1.0  # Exploration rate
        self.epsilon_min = 0.01
        self.epsilon_decay = 0.995
        self.learning_rate = 0.001
        self.model = self._build_model()

    def _build_model(self):
        """Neural Network for Deep Q-Learning"""
        model = Sequential([
            Dense(64, input_dim=self.state_size, activation='relu'),
            Dense(64, activation='relu'),
            Dense(self.action_size, activation='linear')  # Output Q-values for actions
        ])
        model.compile(loss='mse', optimizer=keras.optimizers.Adam(learning_rate=self.learning_rate))
        return model

    def remember(self, state, action, reward, next_state, done):
        """Store experience in memory"""
        self.memory.append((state, action, reward, next_state, done))

    def act(self, state):
        """Choose action based on epsilon-greedy strategy"""
        if np.random.rand() <= self.epsilon:
            return random.randrange(self.action_size)  # Random action
        q_values = self.model.predict(state)
        return np.argmax(q_values[0])  # Choose best action

    def replay(self, batch_size):
        """Train the model using replay memory"""
        minibatch = random.sample(self.memory, batch_size)
        for state, action, reward, next_state, done in minibatch:
            target = reward
            if not done:
                target += self.gamma * np.amax(self.model.predict(next_state)[0])
            target_f = self.model.predict(state)
            target_f[0][action] = target
            self.model.fit(state, target_f, epochs=1, verbose=0)
        if self.epsilon > self.epsilon_min:
            self.epsilon *= self.epsilon_decay  # Decay epsilon

```

```python
import numpy as np
import tensorflow as tf
from sklearn.preprocessing import LabelEncoder

# Drop missing values
df.dropna(inplace=True)

# Convert Categorical Variables to Numeric
label_encoders = {}
for col in df.select_dtypes(include=['object']).columns:
    label_encoders[col] = LabelEncoder()
    df[col] = label_encoders[col].fit_transform(df[col])  # Convert categorical to numeric

# Convert all numeric data to float32 to avoid dtype issues
df = df.astype(np.float32)

# ✅ Ensure CustomerRetentionEnv is imported and defined before initializing
try:
    env = CustomerRetentionEnv(df)  # Initialize Environment
    print("Environment initialized successfully!")
except NameError:
    print("Error: CustomerRetentionEnv is not defined. Make sure to import or define it.")

# ✅ Ensure DQN_Agent is imported and defined before initializing
try:
    agent = DQN_Agent(state_size=env.state_size, action_size=env.action_size)  # Initialize Agent
    print("Agent initialized successfully!")
except NameError:
    print("Error: DQN_Agent is not defined. Make sure to import or define it.")

# Check for GPU acceleration
print("Num GPUs Available:", len(tf.config.experimental.list_physical_devices('GPU')))

# Set GPU usage for faster processing
gpus = tf.config.experimental.list_physical_devices('GPU')
if gpus:
    try:
        tf.config.experimental.set_memory_growth(gpus[0], True)  # Prevents memory overflow
    except RuntimeError as e:
        print(e)

# Reduce total episodes to speed up training
EPISODES = 50

for e in range(EPISODES):
    state = env.reset()
    state = np.reshape(state, [1, env.state_size])
    done = False
    total_reward = 0
    step_count = 0  # Track steps per episode

    while not done:
        action = agent.act(state)  # ✅ Fix: Ensure agent is defined
        next_state, reward, done, _ = env.step(action)
        next_state = np.reshape(next_state, [1, env.state_size])
        agent.remember(state, action, reward, next_state, done)
        state = next_state
        total_reward += reward

        step_count += 1
        if step_count > 500:  # Stop overly long episodes
            done = True

    print(f"Episode {e+1}/{EPISODES}, Reward: {total_reward}")

    # Train model less frequently to save time
    if len(agent.memory) > 32 and e % 5 == 0:
        agent.replay(16)  # Reduce batch size for speed
```
